```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import plotly.express as px
```

```python
df = pd.read_csv("/Users/rounaksarkar/Downloads/deskTop/ML:AI/5. NLP Projects/Duplicate-Question-Pair_Detector/train.csv")
df.shape
```

```python
df.head()
```

```python
new_df = df.sample(30000)
print (new_df.isnull().sum())
```

```python
new_df.duplicated().sum()
```

```python
question_df = new_df[['question1', 'question2']]
question_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
questions = list(question_df['question1']) + list(question_df['question2'])
questions = [str(q) if not pd.isnull(q) else '' for q in questions]  # Replace NaN values with empty string
cv = CountVectorizer(max_features=3000)
q1_arr, q2_arr = np.vsplit(cv.fit_transform(questions).toarray(),2)
```

```python
temp_df1 = pd.DataFrame(q1_arr, index=question_df.index)
temp_df2 = pd.DataFrame(q2_arr, index=question_df.index)
temp_df = pd.concat([temp_df1, temp_df2], axis=1)
temp_df.shape
```

```python
temp_df
```

```python
temp_df['is_duplicate'] = new_df['is_duplicate']
temp_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(temp_df.iloc[:, 0:-1].values, temp_df.iloc[:,-1].values, test_size=0.2, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
y_pred = xgb.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```
